Matrix multiplication involves multiplying each element of a row in the first matrix by the corresponding element of a column in the second matrix and summing up the results to get the elements of the resulting matrix. For a matrix multiplication between an (n, k) matrix (A) and a (k, m) matrix (B), the resulting matrix will be (n, m).

To compute each element in the resulting matrix, you need to perform k multiplications and (k-1) additions. This is because you multiply each element of the corresponding row in the first matrix by the element in the corresponding column of the second matrix, and then sum up these products.

Question-2

```python
def matrix_multiply_lists(mat1, mat2):
    # Get the dimensions of the matrices
    n = len(mat1)
    k = len(mat1[0])
    m = len(mat2[0])

    # Initialize the result matrix with zeros
    result = [[0 for l in range(m)] for l in range(n)]

    # Perform matrix multiplication
    for i in range(n):
        for j in range(m):
            for l in range(k):
                result[i][j] += mat1[i][l] * mat2[l][j]

    return result


# Numpy_Approach

import numpy as np

def matrix_multiply_numpy(mat1, mat2):
    return np.dot(mat1, mat2)

# Comparing the time taken for both the approaches:

import timeit

# Example matrices

mat1 = [[1, 2], [3, 4]]
mat2 = [[5, 6], [7, 8]]

# Timing for list of lists approach
time_lists = timeit.timeit(lambda: matrix_multiply_lists(mat1, mat2), number=1000)

# Timing for NumPy approach
np_mat1 = np.array(mat1)
np_mat2 = np.array(mat2)
time_numpy = timeit.timeit(lambda: matrix_multiply_numpy(np_mat1, np_mat2), number=1000)

print(f"Time for list of lists approach: {time_lists} seconds")
print(f"Time for NumPy approach: {time_numpy} seconds")
```

We can see that the matrix multiplication is faster using numpy

```python
# A better method to find the median of the array is as follows:

def find_median_sorted(arr):
    sorted_arr = sorted(arr)
    n = len(sorted_arr)

    # If the size of the array is even, take the average of the two middle elements
    if n % 2 == 0:
        return (sorted_arr[n // 2 - 1] + sorted_arr[n // 2]) / 2
    else:
        return sorted_arr[n // 2]

# Numpy Approach

import numpy as np

def find_median(arr):
    return np.median(arr)

# Comparing the time complexity of both approaches:

import timeit

# Example array
arr = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]

# Timing for finding median using the sorting method
time_median_sorted = timeit.timeit(lambda: find_median_sorted(arr), number=1000)

# Timing for NumPy's median
time_numpy_median = timeit.timeit(lambda: find_median(arr), number=1000)

print(f"Time for sorted method: {time_median_sorted} seconds")
print(f"Time for NumPy's median: {time_numpy_median} seconds")
```

```python
class Student:
    def __init__(self, roll_number, name, marks):
        self.roll_number = roll_number
        self.name = name
        self.marks = marks  # Dictionary to store subject-wise marks

class Branch:
    def __init__(self, name):
        self.name = name
        self.students = []  # List to store student objects

    def add_student(self, student):
        self.students.append(student)

class Year:
    def __init__(self, year):
        self.year = year
        self.branches = []  # List to store branch objects

    def add_branch(self, branch):
        self.branches.append(branch)

# Creating instances for the example
student1 = Student(1, 'N', {'Maths': 100, 'English': 70})
student2 = Student(2, 'M', {'Maths': 90, 'English': 80})

branch1 = Branch('Branch 1')
branch1.add_student(student1)

branch2 = Branch('Branch 2')
branch2.add_student(student2)

year2022 = Year(2022)
year2022.add_branch(branch1)
year2022.add_branch(branch2)

print(year2022.branches[0].students[0].roll_number)
print(year2022.branches[0].students[0].name)
print(year2022.branches[0].students[0].marks)
```

```python
import numpy as np
import pandas as pd

np.random.seed(42)

# Generating a 20x5 matrix of random numbers uniformly drawn from the range [1, 2)
random_matrix = np.random.uniform(1, 2, size=(20, 5))

# Creating a DataFrame from the matrix with column names "a", "b", "c", "d", "e"
df = pd.DataFrame(random_matrix, columns=["a", "b", "c", "d", "e"])

# Add a new column "f" as the sum of columns "a", "b", "c", "d", "e"
df['f'] = df[['a', 'b', 'c', 'd', 'e']].sum(axis=1)

# Create a new column "g" based on the value in the column "f"
df['g'] = np.where(df['f'] < 8, 'LT8', 'GT8')

# Find the number of rows where the value in the column "g" is "LT8"
num_lt8_rows = (df['g'] == 'LT8').sum()

# Finding the standard deviation of the column "f" for rows where "g" is "LT8" and "GT8"
std_lt8 = df.loc[df['g'] == 'LT8', 'f'].std()
std_gt8 = df.loc[df['g'] == 'GT8', 'f'].std()

print("Updated DataFrame:")
print(df)
print("\nNumber of rows where the value in the column 'g' is 'LT8':", num_lt8_rows)
print("Standard deviation of 'f' for rows where 'g' is 'LT8':", std_lt8)
print("Standard deviation of 'f' for rows where 'g' is 'GT8':", std_gt8)

```

```python
import numpy as np

# Create a 2x3 array
arr1 = np.array([[1, 2, 3],
                 [4, 5, 6]])

# Create a 1x3 array
arr2 = np.array([10, 20, 30])

# Perform broadcasting to add arr2 to each row of arr1
result = arr1 + arr2

# Display the original arrays and the result
print("Array 1:")
print(arr1)
print("\nArray 2:")
print(arr2)
print("\nArray after Broadcasting:")
print(result)
```

In this example, arr1 is a 2x3 array, and arr2 is a 1x3 array. Normally, you wouldn't be able to add them directly because their shapes are different. However, broadcasting allows NumPy to implicitly expand the dimensions of the smaller array to match the shape of the larger array during the operation.
Here, NumPy broadcasts the addition operation, adding arr2 to each row of arr1 element-wise.

```python
import numpy as np

matrix = np.random.uniform(0, 10, size=(3, 3))
print("Original Matrix:")
print(f"Matrix:\n{matrix}\n")

def find_custom_argmin(arr):
    flat_arr = arr.ravel()
    min_index = 0

    for index, value in enumerate(flat_arr):
        if value < flat_arr[min_index]:
            min_index = index

    return min_index

custom_argmin_result = find_custom_argmin(matrix)
print(f"Custom Argmin Result: {custom_argmin_result}\n")

numpy_argmin_result = np.argmin(matrix)
print(f"NumPy Argmin Result: {numpy_argmin_result}")
```
